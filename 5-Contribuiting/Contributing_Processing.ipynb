```python
from datetime import datetime
import os
import pandas as pd
os.getcwd()
import pandas as pd
import sys
import numpy
from difflib import SequenceMatcher

import nltk
#import nltk .download()
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud
import string
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
stopwords = stopwords.words('english')
```

```python
google = pd.read_csv("../5-Contributing/googleContributingOnly.csv")
aws = pd.read_csv("../5-Contributing/awsContributingOnly.csv")
azure = pd.read_csv("../5-Contributing/azureContributingOnly.csv")
spring = pd.read_csv("../5-Contributing/springContributingOnly.csv")
```

```python
googleIndex = len(google['fileContribuiting'])
awsIndex = len(aws['fileContribuiting'])
azureIndex = len(azure['fileContribuiting'])
springIndex = len(spring['fileContribuiting'])

```

```python
def exportar(file,framework):
  text_file = open("5-Contributing/"+framework +"ContributingUnion.txt", "w")
  n = text_file.write(file)
  text_file.close()
```

```python
azureJoin =  " ".join(azure['fileContribuiting'])
awsJoin =  " ".join(aws['fileContribuiting'])
googleJoin =  " ".join(google['fileContribuiting'])
springJoin =  " ".join(spring['fileContribuiting'])

azureJoin = azureJoin.lower()
googleJoin = googleJoin.lower()
springJoin = springJoin.lower()
awsJoin = awsJoin.lower()

exportar(azureJoin,'azure')
exportar(googleJoin,'google')
exportar(springJoin,'spring')
exportar(awsJoin,'aws')
```

```python
googleContributingUnion = open("../5-Contributing/googleContributingUnion.txt", mode = 'r', encoding= 'utf-8').read()
awsContributingUnion = open("../5-Contributing/awsContributingUnion.txt", mode = 'r', encoding= 'utf-8').read()
azureContributingUnion = open("../5-Contributing/azureContributingUnion.txt", mode = 'r', encoding= 'utf-8').read()
springContributingUnion = open("../5-Contributing/springContributingUnion.txt", mode = 'r', encoding= 'utf-8').read()
```

```python
print(googleJoin)
```

```python
text_s_pontucao_google = ''.join([p for p in googleJoin if p not in string.punctuation])
print(text_s_pontucao_google)
```

```python
tokenize_word_google = nltk.word_tokenize(text_s_pontucao_google)

print(tokenize_word_google)
```

```python
#remoção das stopwords palavras a qual quando retiradas do text não faz o texto perder o sentido

#nltk.download('stopwords')
#stopwords = stopwords.words('english')

word_google_s_stop = [p for p in tokenize_word_google if p not in stopwords]
print(word_google_s_stop)
```

```python
freq_google = FreqDist(word_google_s_stop)
freq_google = freq_google.most_common(50)
print(freq_google)
```

```python
valueMediaSimilar = []
for i in range(azureIndex):
    doc1 = azure['fileContribuiting'][i]
    valueSimilar = []
    for j in range(azureIndex):

      doc2 = azure['fileContribuiting'][j]
      if(i != j):
        valueSimilar.append(SequenceMatcher(None, doc1, doc2).ratio()) # lambda x: x in "# \t\n"
    valueMediaSimilar.append(numpy.mean(valueSimilar))
    #print(numpy.mean(valueSimilar))

print(numpy.median(valueMediaSimilar))
```
