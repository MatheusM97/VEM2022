```python
def remove_internal_pull_requests(pulls, maintainers):
  #unique_maintainers = maintainers["login"].unique()
  #print(len(maintainers))
  #print(maintainers)
  last_index = pulls.index[-1]
  for index in pulls.index:
   # print(index, last_index)
    line = pulls.loc[index]
    creator = line["creator user login"]
    if creator in maintainers:
      pulls.drop(index, inplace=True)
```

```python
def remove_auto_closed(pulls):
  last_index = pulls.index[-1]
  for index in pulls.index:
   # print(index, last_index)
    line = pulls.loc[index]
    if (line["creator user login"] == line["closed user login"]):
      pulls.drop(index, inplace=True)
```

```python
remove_auto_closed(google)
remove_auto_closed(spring)
remove_auto_closed(aws)
remove_auto_closed(azure)
```

```python
from numpy import NaN
def agregarColaboradores(framework_PRs,framework, logins):
  objetos = list()
  numero_prs_criados = 0
  numero_merges = 0
  numero_prs_criado_mergeado = 0
  numero_prs_criado_n_mergeado = 0
  numero_prs_nmergeado_fechado = 0
  numero_prs_fechado = 0
  seguidores = 0

  data_inicial = 0
  data_final = 0
  valor_dias_experiencia = 0

  framework_PRs.columns = framework_PRs.columns.str.replace(' ', '_')
  #logins = framework_PRs[(framework_PRs['merged'] == True)]['merge_user_login'].dropna().unique()
  print(len(logins))
  print(logins)
  for i in range(0,len(logins)):
    print(logins[i])
    #merged by maintainer independent contributor
    numero_merges = framework_PRs[(framework_PRs.merge_user_login == logins[i])]['merge_user_login'].count()
    #prs created by maintainer
    numero_prs_criados = framework_PRs[(framework_PRs.creator_user_login == logins[i])]['creator_user_login'].count()
    #creator and merged by maintainer
    numero_prs_criado_mergeado = framework_PRs[(framework_PRs.creator_user_login == logins[i]) & (framework_PRs.merge_user_login == logins[i])]['merge_user_login'].count()
    #created by maintainer but not merged and closed by someone
    numero_prs_criado_n_mergeado = framework_PRs[(framework_PRs.creator_user_login == logins[i]) & (framework_PRs.merged == False) & (framework_PRs.status == 'closed')]['creator_user_login'].count()
    #created by anyone, not merged and closed by the maintainer in question
    numero_prs_nmergeado_fechado = framework_PRs[(framework_PRs.closed_user_login == logins[i]) & (framework_PRs.merged == False)  & (framework_PRs.status == 'closed')]['closed_user_login'].count()
    #created by anyone, independent of merged and closed by the maintainer
    numero_prs_fechado = framework_PRs[(framework_PRs.closed_user_login == logins[i])  & (framework_PRs.status == 'closed')]['closed_user_login'].count()
    #created by anyone(not maintainer) and merged
    numero_merges_exclude_auto_merge = framework_PRs[(framework_PRs.merge_user_login == logins[i]) & (framework_PRs.creator_user_login != logins[i])]['merge_user_login'].count()
    #created by anyone(not maintainer) and just closed (reviewed accepted or not)
    numero_revisoes_exclude_auto_revisado = framework_PRs[(framework_PRs.closed_user_login == logins[i])  & (framework_PRs.status == 'closed') & (framework_PRs.creator_user_login != logins[i])]['closed_user_login'].count()
    #auto_review
    auto_revisado = framework_PRs[(framework_PRs.closed_user_login == logins[i])  & (framework_PRs.status == 'closed') & (framework_PRs.creator_user_login == logins[i])]['closed_user_login'].count()

    #print(numero_prs_nmergeado_fechado)
    #print(numero_merges)
    #if exists closed, but don't merged
    if(numero_merges > 0 or numero_prs_nmergeado_fechado > 0):
      seguidores = framework_PRs[(framework_PRs.closed_user_login == logins[i])]['closed_user_seguidores'].dropna().sort_values()
      #print(seguidores)
      if(len(seguidores) > 0):
        seguidores = seguidores.iloc[-1]

      #print("seguidores " + str(seguidores))
      datas = framework_PRs[(framework_PRs.merge_user_login == logins[i])]['data_merge'].dropna().sort_values()
      datas_closed = framework_PRs[(framework_PRs.closed_user_login == logins[i]) ]['data_de_fechamento'].dropna().sort_values()
      #print(datas)
      #print(datas_closed)

      if(numero_merges_exclude_auto_merge > 0):
        if datas.iloc[0] > datas_closed.iloc[0]:
          data_inicial = datas_closed.iloc[0]
        else:
          data_inicial =  datas.iloc[0]

        #print(data_inicial)

        if datas.iloc[-1] < datas_closed.iloc[-1]:
          data_final = datas_closed.iloc[-1]
        else:
          data_final =  datas.iloc[-1]
        #print( data_final)


        valor_dias_experiencia =  (datetime.strptime(data_final, '%Y-%m-%d %H:%M:%S') - datetime.strptime(data_inicial, '%Y-%m-%d %H:%M:%S')).days
        #print(valor_dias_experiencia)
    else:
      seguidores = ''
      datas = framework_PRs[(framework_PRs.merge_user_login == logins[i])]['data_merge'].dropna().sort_values()
      datas_closed = framework_PRs[(framework_PRs.closed_user_login == logins[i]) ]['data_de_fechamento'].dropna().sort_values()


    #print(datas)
    #print(datas_closed)
    #print(numero_prs_criados)
    #print(numero_merges)
    #print(numero_prs_criado_mergeado)
    #print(numero_prs_criado_n_mergeado)
    #print(numero_prs_nmergeado_fechado)
    #print(numero_prs_fechado)
    #print(numero_merges_exclude_auto_merge)



    objeto = {
        "framework": framework,
        "login": logins[i],
        "seguidores": seguidores,
        "numero_prs_criados":  numero_prs_criados,
        "numero_merges":  numero_merges,
        "numero_prs_criado_mergeado": numero_prs_criado_mergeado,
        "numero_prs_criado_n_mergeado": numero_prs_criado_n_mergeado,
        "numero_prs_n_mergeado_fechado": numero_prs_nmergeado_fechado,
        "numero_prs_fechado": numero_prs_fechado,
        "numero_merges_exclude_auto_merge" : numero_merges_exclude_auto_merge,
        "numero_revisoes_exclude_auto_revisado": numero_revisoes_exclude_auto_revisado,
        "auto_revisado" : auto_revisado,
        "tempo experiencia(dias)": valor_dias_experiencia
    }
    objetos.append(objeto)

  df = pd.DataFrame(objetos)
  return df
```
